# 🧠 EverLight Sync Engine
This notebook scans the `/user_updates/` folder for `.md` or `.txt` files,
logs their contents into a dated `.md` file inside `/logs/`, and reports how many updates it processed.

Run this anytime to sync new memory entries into your local archive.

```python
from pathlib import Path
import datetime

# Define folders
update_folder = Path("../user_updates")
log_folder = Path("../logs")
log_folder.mkdir(parents=True, exist_ok=True)

# Create log file
log_file = log_folder / f"log_{datetime.datetime.now().strftime('%Y-%m-%d')}.md"

entries = []

# Read each file in user_updates
for file in update_folder.glob("*"):
    if file.suffix in [".md", ".txt"]:
        content = file.read_text(encoding="utf-8")
        entries.append(f"## {file.name}\n\n{content}\n\n---\n")

# Write to log
if entries:
    log_file.write_text("\n".join(entries), encoding="utf-8")
    print(f"✅ Synced {len(entries)} entries to {log_file.name}")
else:
    print("📭 No new updates found in user_updates.")
```
