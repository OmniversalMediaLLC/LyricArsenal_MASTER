```python
import datetime
from pathlib import Path

update_folder = Path("../user_updates")
log_folder = Path("../logs")
log_file = log_folder / f"log_{datetime.datetime.now().strftime('%Y-%m-%d')}.md"

entries = []

for file in update_folder.glob("*"):
    if file.suffix in [".md", ".txt"]:
        content = file.read_text()
        entries.append(f"## {file.name}\n\n{content}\n---\n")

if entries:
    log_file.write_text("\n".join(entries))
    print(f"âœ… Synced {len(entries)} entries to {log_file.name}")
else:
    print("ðŸ“­ No new updates found.")
```
